```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def display_info(self):
        print("Name:", self.name)
        print("Salary:", self.salary)

    def work(self):
        print(self.name, "is working.")

class Teacher(Employee):
    def __init__(self, name, salary, subject):
        super().__init__(name, salary)
        self.subject = subject

    def display_info(self):
        super().display_info()
        print("Subject:", self.subject)

    def work(self):
        print(self.name, "is teaching", self.subject)

class Developer(Employee):
    def __init__(self, name, salary, language):
        super().__init__(name, salary)
        self.language = language

    def display_info(self):
        super().display_info()
        print("Programming Language:", self.language)

    def work(self):
        print(self.name, "is coding in", self.language)

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department

    def display_info(self):
        super().display_info()
        print("Department:", self.department)

    def work(self):
        print(self.name, "is managing", self.department)
```

```python
teacher = Teacher("Ali", 50000, "Physics")
developer = Developer("Ahmed", 80000, "Python")
manager = Manager("Sara", 100000, "HR")

print("------ Teacher ------")
teacher.display_info()
teacher.work()

print()

print("------ Developer ------")
developer.display_info()
developer.work()

print()

print("------ Manager ------")
manager.display_info()
manager.work()
```
