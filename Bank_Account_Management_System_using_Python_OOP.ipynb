```python
class BankAccount:
    def __init__(self, amount):
        if amount < 0:
            raise ValueError("Initial amount cannot be negative")
        self._amount = amount

    # Getter using @property
    @property
    def amount(self):
        return self._amount

    # Setter using @amount.setter
    @amount.setter
    def amount(self, value):
        if value < 0:
            raise ValueError("Amount cannot be negative")
        self._amount = value

    # Deposit Method
    def deposit(self, money):
        if money <= 0:
            print("Deposit amount must be greater than 0.")
        else:
            self._amount += money
            print(f"Rs.{money} deposited successfully.")

    # Withdraw Method
    def withdraw(self, money):
        if money <= 0:
            print("Withdrawal amount must be greater than 0.")
        elif money > self._amount:
            print("Insufficient Balance!")
        else:
            self._amount -= money
            print(f"Rs.{money} withdrawn successfully.")
```

```python
# Create Account
account = BankAccount(1000)

# Getter
print("Current Balance:", account.amount)

# Deposit
account.deposit(500)
print("Balance after Deposit:", account.amount)

# Withdraw
account.withdraw(300)
print("Balance after Withdrawal:", account.amount)

# Setter
account.amount = 2000
print("Updated Balance:", account.amount)
```
